## Detect if website has tenders

```python
gov_list = ['https://www.gld.gov.hk/egazette/english/search_gazette/search_gov_notice.html',
'https://www.cedb.gov.hk/ccib/eng/tender/index.htm',
'https://www.edb.gov.hk/en/public-admin/tender-notices/tender-notice/index.html',
'https://www.itb.gov.hk/en/tender/',
'https://www.itc.gov.hk/en/tender/tender.htm',
'http://www.aud.gov.hk/eng/tender/tender.htm',
'http://www.ams.gov.hk/eng/tender1.htm',
'https://www.admwing.gov.hk/eng/tender/tender.htm',
'http://www.cas.gov.hk/eng/notice/notice_tender.html',
'https://www.cad.gov.hk/english/tender.html',
'https://www.csb.gov.hk/english/tender/17.html',
'https://www.cedb.gov.hk/citb/en/Tender_Notices/index.html',
'https://www.cr.gov.hk/en/tender/index.htm',
'http://www.cmab.gov.hk/en/tender/index.htm',
# 'https://www.csd.gov.hk/english/tender/other_tender.html', really long title
'https://www.customs.gov.hk/en/customs_notices/tender/index.html',
'http://www.dh.gov.hk/english/tender/tender.html',
'https://www.doj.gov.hk/eng/tender/',
'http://www.effo.gov.hk/en/news-events/tender-notice.html',
'https://www.enb.gov.hk/en/susdev/tender/index.htm',
'https://www.fstb.gov.hk/tb/en/tender-exercises.htm',
'https://www.fstb.gov.hk/fsb/tender/index.htm',
'http://www.hkfsd.gov.hk/eng/tender/tender_notice/index.html',
'http://www.fhb.gov.hk/en/tender/',
'http://www.gfs.gov.hk/eng/tender.htm',
# 'http://www.hkma.gov.hk/eng/about-the-hkma/tender-invitations.shtml', tender names do not have standard words
'http://www.hko.gov.hk/tender/tender.htm',
'https://www.police.gov.hk/ppp_en/11_useful_info/tenders.html',
'https://www.hongkongpost.hk/en/about_us/tender/notice/index.html',
'http://www.icac.org.hk/en/so/tender/index.html',
'https://www.ird.gov.hk/eng/ten/ten.htm',
'http://www.try.gov.hk/internet/ehtender.html',
'https://www.tid.gov.hk/english/aboutus/noticeboard/tender.html',]
```

```python
link_list = []
status_list = []
tender_name_list = []

for link in gov_list:
    page = requests.get(link)
    status = page.status_code
    
    
    if status == 200:
        soup = BeautifulSoup(page.content, 'html.parser')
        sentences = soup.text.split('\n')
        sentences = list(filter(None, sentences))
        sentences = list(filter(lambda space: space.strip(), sentences))
        
        for sentence in sentences:
            if ('Provision' in sentence and 'Provisions' not in sentence) or 'Supply' in sentence or 'Quotation' in sentence:
                link_list.append(link)
                status_list.append('Received')
                tender_name_list.append(sentence)
```

```python
csd_url = 'https://www.csd.gov.hk/english/tender/other_tender.html'
page = requests.get(csd_url)
soup = BeautifulSoup(page.content,'html.parser')
tables = soup.find_all('tbody')[1:3]

for table in tables:
    sentences = table.get_text().split('\t')
    sentences = list(filter(None, sentences))
    sentences = list(filter(lambda name: name.strip(), sentences))
    
    for sentence in sentences:
        if ('Provision' in sentence and 'Provisions' not in sentence) or 'Supply' in sentence or 'Quotation' in sentence:
            link_list.append(csd_url)
            status_list.append('Received')
            tender_name_list.append(sentence)
```

```python
hkma_url = 'http://www.hkma.gov.hk/eng/about-the-hkma/tender-invitations.shtml'
page = requests.get(hkma_url)
soup = BeautifulSoup(page.content,'html.parser')
item = soup.find('div', class_='item')
tenders = item.find_all('a')
for tender in tenders:
    link_list.append(hkma_url)
    status_list.append('Received')
    tender_name_list.append("Provision of " + tender.get_text())
```

```python
d = {'links': link_list,
     'status': status_list,
     'tender_name': tender_name_list,
    }

df = pd.DataFrame(data=d)
df = df[['links', 'status', 'tender_name']]
```

```python
def clean_name(row):
    name_parts = row.tender_name.split(' ')
    for i in name_parts:
        if i == 'Provision' or i == 'Supply' or i == 'Quotation':
            start = name_parts.index(i)
    clean = name_parts[start:]
    together = ' '.join(clean)
    return together
```

```python
df['tender_name'] = df.apply(clean_name, axis=1)
```

```python
digital_terms = ['software', 'digital', 'comput', 'system', 'mobile', 
                 'web', 'consult', 'cms', 'network', 'infrastructure']

def digital(row):
    for term in digital_terms:
        if term in row.tender_name.lower():
            return True
    return False

exact_digital_terms = ['app', 'apps', 'it']

def exact_words(row):
    name = row.tender_name.lower()
    parts = name.split()
    for term in exact_digital_terms:
        for part in parts:
            if term == part:
                return True
    return row.digital
```

```python
df['digital'] = df.apply(digital, axis=1)
df['digital'] = df.apply(exact_words, axis=1)
```
